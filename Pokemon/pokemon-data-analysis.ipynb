# Pokemon Data Analysis

This Notebook provides Data Analysios of A CSV File Containing the Information and Data about different types of Pokemon.

First of All Importing all libraries that will be used in Analysis.

```python
import pandas as pandasInstance
import numpy as numpyInstance
import seaborn as seabornInstance
import matplotlib.pyplot as matplotlibInstance
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
```

Now Doing the Settings for Offline Graphs and Data Visualizations.

```python
%matplotlib inline
init_notebook_mode(connected=True)
cf.go_offline()
```

Now importing data of Poekemons from the File.

```python
pokemonDataFrame = pandasInstance.read_csv('../input/Pokemon.csv')
```

Now checking the Header of the file.

```python
pokemonDataFrame.head()
```

Now checking the information of each column.

```python
pokemonDataFrame.info()
```

Now if we want to visulaize the Total Number of Different types of Pokemon.

```python
matplotlibInstance.figure(figsize=(30,25))
matplotlibInstance.tight_layout()
seabornInstance.set(font_scale=2)
seabornInstance.countplot(x='Type 1',data=pokemonDataFrame)
```

Now if want a comparison of Legendary and Non-Legendary Pokemon we can do that by a pie chart.

```python
byLegendary = pokemonDataFrame.groupby(by='Legendary').sum()
```

```python
byLegendaryTotal = byLegendary['Total']
```

```python
matplotlibInstance.figure(figsize=(10,30))
matplotlibInstance.pie(byLegendaryTotal,labels=byLegendaryTotal.index,shadow=True)
```

```python
matplotlibInstance.figure(figsize=(30,25))
matplotlibInstance.tight_layout()
pokemonDataFrame.iplot(kind='bar',x='Type 1',y='Total',color='purple')
```

Now if we want to check the HP of Each Pokemon Correspodnig to its Name we can Make a Line Graph.

```python
matplotlibInstance.figure(figsize=(30,25))
matplotlibInstance.tight_layout()
pokemonDataFrame[['Name','HP']].iplot()
```

Here we can See that Name is increasing it because moving forward the Number of Name Keep on increasing.

If we want to check the Number of Different Pokemons of the Different Generations.

```python
matplotlibInstance.figure(figsize=(30,25))
matplotlibInstance.tight_layout()
seabornInstance.countplot(x='Generation',data=pokemonDataFrame)
```

Now if we want to categorize the Pokemon on the basis of the Attacking Power.

```python
matplotlibInstance.figure(figsize=(30,25))
matplotlibInstance.tight_layout()
pokemonDataFrame.iplot(kind='bar',x='Type 1',y='Attack',color='red')
```

```python
matplotlibInstance.figure(figsize=(30,25))
matplotlibInstance.tight_layout()
seabornInstance.boxplot(x='Type 1',y='Speed',data=pokemonDataFrame)
```
